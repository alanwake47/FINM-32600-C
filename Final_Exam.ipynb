#### Difference between static and non static members in C++

In C++, a class can have both static and non-static member variables and member functions. Here are the differences between them:

Scope: Non-static members have an instance-level scope, meaning that each object of the class has its own copy of the non-static member variables. On the other hand, static members have a class-level scope, meaning that there is only one copy of the static member variables shared by all objects of the class.

Memory allocation: Non-static members are allocated memory for each object of the class, while static members are allocated memory only once, regardless of how many objects of the class are created.

Initialization: Non-static members can be initialized in the constructor of the class, while static members are initialized outside the class definition.

Access: Non-static members can be accessed using the dot operator on an object of the class, while static members can be accessed using the scope resolution operator on the class itself.

Here's an example that demonstrates the differences between static and non-static members:

Since nonStaticVar is a non-static member variable, each object of the class has its own copy, and we set the value of nonStaticVar separately for obj1 and obj2. On the other hand, staticVar is a static member variable, and there is only one copy of it shared by all objects of the class.

### Lazy initialization 

is a design pattern in C++ that defers the initialization of an object or variable until it is actually needed. This can be useful in situations where the initialization of an object is expensive or time-consuming, and you want to avoid doing it until it is absolutely necessary.

In lazy initialization, you typically create a placeholder object or variable that represents the uninitialized state of the object. When the object is actually needed, you initialize it on demand and replace the placeholder with the fully initialized object. This allows you to avoid the cost of initialization until it is actually required, improving performance and reducing memory usage.

One common way to implement lazy initialization in C++ is to use a static local variable in a function. For example, consider the following code:

MyClass& getMyClassInstance() { \
    static MyClass instance; \
    return instance; \
} 

In this code, getMyClassInstance is a function that returns a reference to a static local variable instance of type MyClass. The first time getMyClassInstance is called, the instance variable is initialized with the default constructor for MyClass. On subsequent calls, the already-initialized instance variable is returned.

Another way to implement lazy initialization is to use the singleton design pattern. In this pattern, you create a class with a private constructor and a static method that returns a reference to a static instance of the class. The first time the static method is called, the instance is created using the private constructor, and on subsequent calls, the already-created instance is returned.

#### Smart Pointers
A Shared Pointer is a smart pointer that retains shared ownership of an object through a pointer. Several shared pointers may own the same object.

In C++, a smart pointer is a class template that provides a way to manage dynamically allocated memory. A smart pointer acts like a regular pointer, but it also provides additional features like automatic memory management, preventing memory leaks, and ensuring that memory is freed when it's no longer needed.

The primary advantage of using smart pointers is that they can help prevent common programming mistakes like memory leaks or dangling pointers. Smart pointers do this by automatically managing the lifetime of an object and freeing its memory when it's no longer needed.

In C++, there are three main types of smart pointers available in the standard library:

std::unique_ptr: A smart pointer that manages the lifetime of a single object. It ensures that only one pointer owns the object, and when that pointer is destroyed, the object is automatically deleted.

std::shared_ptr: A smart pointer that manages the lifetime of an object that might have multiple pointers pointing to it. It uses a reference counting mechanism to ensure that the object is deleted only when all of the pointers pointing to it are destroyed.

std::weak_ptr: A weak reference to a std::shared_ptr that doesn't increase the reference count. It's used to prevent circular references between std::shared_ptrs, and it can be used to check if the object is still alive.

Overall, smart pointers are an essential feature of modern C++ programming, and they help ensure that your code is safer and more reliable.

    // Use the smart pointer to access the integer
    std::cout << *ptr << std::endl;  // Output: 10

    // Change the value of the integer
    *ptr = 20;

    // Use the smart pointer to access the new value of the integer
    std::cout << *ptr << std::endl;  // Output: 20

    // The memory managed by the smart pointer is automatically deallocated when it goes out of scope
    return 0;
}

In this example, we use the std::unique_ptr class to create a smart pointer to an integer with an initial value of 10. We then use the pointer to access the value of the integer and change its value to 20. Finally, when the smart pointer goes out of scope, the memory it manages is automatically deallocated.

Note that there are other types of smart pointers available in C++, such as std::shared_ptr and std::weak_ptr, which provide different levels of ownership and reference counting semantics. The appropriate type of smart pointer to use depends on the specific requirements of the program.

##### Virtual keyword
The virtual keyword is used to override a function in a derived class. When we override a function, we are redefining the function in the derived class. The virtual keyword is used to tell the compiler that we are overriding a function.

#### Auto keyword

The auto keyword is used to declare a variable whose type is automatically deduced from its initializer.

In C++, an abstract class is a class that has at least one pure virtual function. A pure virtual function is a virtual function that is declared using the = 0 syntax, and has no implementation in the base class.

An abstract class cannot be instantiated directly; it can only be used as a base class for other classes. Any class that inherits from an abstract class must provide an implementation for all the pure virtual functions in the base class, otherwise it will also be considered an abstract class and cannot be instantiated.

Here is an example of an abstract class in C++:

class Shape {
public:
    virtual double area() const = 0;
    virtual double perimeter() const = 0;
};

class Rectangle : public Shape {
public:
    Rectangle(double w, double h) : width(w), height(h) {}

    double area() const override { return width * height; }
    double perimeter() const override { return 2 * (width + height); }

private:
    double width;
    double height;
};

class Circle : public Shape {
public:
    Circle(double r) : radius(r) {}

    double area() const override { return 3.14159 * radius * radius; }
    double perimeter() const override { return 2 * 3.14159 * radius; }

private:
    double radius;
};

In this example, Shape is an abstract class that defines two pure virtual functions: area() and perimeter(). Any class that inherits from Shape must implement both of these functions.

The Rectangle and Circle classes inherit from Shape and provide their own implementations of the area() and perimeter() functions. Because they provide implementations for all the pure virtual functions in Shape, they are not abstract classes and can be instantiated.

Note that we use the override keyword in the derived classes to indicate that we are overriding the virtual functions from the base class. This helps catch errors at compile time if the function signatures do not match.

#### Polymorphism

Polymorphism is the ability of an object to take on many forms. The most common use of polymorphism in C++ occurs when a parent class reference is used to refer to a child class object.

Polymorphism is the ability of objects of different classes to be treated as if they are objects of the same class. In C++, polymorphism is achieved through two mechanisms: inheritance and virtual functions.

Inheritance allows a derived class to inherit the properties and behavior of its base class. When a derived class inherits from a base class, it can be treated as an object of the base class. This means that any function that takes a pointer or reference to the base class can also take a pointer or reference to the derived class.

Virtual functions allow a function in a base class to be overridden by a function with the same signature in a derived class. When a virtual function is called on an object, the implementation of the function that is called depends on the type of the object at runtime. This means that if a pointer or reference to a base class is used to call a virtual function on a derived class object, the implementation of the virtual function in the derived class will be called.

Here is an example of polymorphism in C++ using inheritance and virtual functions:

class Shape {
public:
    virtual double area() const = 0;
};

class Rectangle : public Shape {
public:
    Rectangle(double w, double h) : width(w), height(h) {}

    double area() const override { return width * height; }

private:
    double width;
    double height;
};

class Circle : public Shape {
public:
    Circle(double r) : radius(r) {}

    double area() const override { return 3.14159 * radius * radius; }

private:
    double radius;
};

int main() {
    std::vector<Shape*> shapes = { new Rectangle(5, 4), new Circle(3) };

    for (auto shape : shapes) {
        std::cout << "Area: " << shape->area() << std::endl;
    }

    for (auto shape : shapes) {
        delete shape;
    }
}

In this example, we define a base class Shape with a pure virtual function area(). We then define two derived classes Rectangle and Circle that inherit from Shape and provide their own implementations of the area() function.

We then create a vector of Shape* pointers that store objects of both Rectangle and Circle classes. We use a for loop to call the area() function on each object in the vector, which is possible because the Shape* pointers can be treated as objects of the Shape class. At runtime, the implementation of the area() function that is called depends on the type of the object pointed to by the Shape* pointer.

Finally, we use another for loop to delete the objects created with new, to avoid memory leaks.

##### CONST Member Functions

In C++, a member function of a class can be declared as const if it does not modify the object it is called on. This means that the this pointer inside the function is treated as a pointer to a constant object, and the function is not allowed to modify the object's state or call non-const member functions.

Declaring a member function as const has several benefits. It makes the code more readable and less error-prone, because it indicates to other programmers that the function will not modify the object it is called on. It also allows objects to be passed as const references or pointers, which can improve performance and reduce memory usage.

Here is an example of a class with a const member function:

class MyClass {
public:
    void setValue(int val) { value = val; }

    int getValue() const { return value; }

private:
    int value;
};

int main() {
    MyClass obj;
    obj.setValue(42);

    const MyClass& constObj = obj;
    std::cout << "Value: " << constObj.getValue() << std::endl;
}

In this example, we define a class MyClass with two member functions: setValue() and getValue(). setValue() is not declared as const, because it modifies the object's state by setting the value of the value member variable. getValue() is declared as const, because it does not modify the object's state, it only returns the value of the value member variable.

In the main() function, we create an object obj of the MyClass class and set its value to 42 using setValue(). We then create a const reference constObj to the same object and call the getValue() function on it. Since getValue() is declared as const, it is allowed to be called on a const reference, and it returns the value of the value member variable without modifying the object's state.

### Virtual Destructor

In C++, when a class is designed to be used as a base class, it is good practice to declare a virtual destructor. This ensures that when an object of the derived class is destroyed, the destructor of the derived class is called as well as the destructor of the base class.

Here is an example that demonstrates the use of a virtual destructor:

class Base {
public:
    virtual ~Base() {}

    // other member functions
};

class Derived : public Base {
public:
    ~Derived() {}

    // other member functions
};

int main() {
    Base* ptr = new Derived();

    // use ptr...

    delete ptr;
}

In this example, we define a base class Base with a virtual destructor and a derived class Derived that inherits from Base. The destructor of Derived is not declared as virtual.

In the main() function, we create a pointer ptr of type Base* that points to a Derived object. We then use the pointer ptr to access the object's member functions. Finally, we delete the object using delete ptr.

Since the destructor of Base is declared as virtual, when delete ptr is called, the destructor of Derived is called first, followed by the destructor of Base. This ensures that all resources allocated by the Derived object are properly deallocated, even if the object is deleted through a pointer to its base class.

If the destructor of Base were not declared as virtual, only the destructor of Base would be called when delete ptr is called, which could lead to memory leaks and undefined behavior.

##### Inheritance

In C++, there are three forms of inheritance: public, private, and protected.

Public inheritance: In public inheritance, the public members of the base class become public members of the derived class, the protected members of the base class become protected members of the derived class, and the private members of the base class remain inaccessible to the derived class.

class Base {
public:
    int publicVar;
protected:
    int protectedVar;
private:
    int privateVar;
};

class Derived : public Base {
    // publicVar is public
    // protectedVar is protected
    // privateVar is not accessible
};


Private inheritance: In private inheritance, the public and protected members of the base class become private members of the derived class, and the private members of the base class remain inaccessible to the derived class.

class Base {
public:
    int publicVar;
protected:
    int protectedVar;
private:
    int privateVar;
};

class Derived : private Base {
    // publicVar is private
    // protectedVar is private
    // privateVar is not accessible
};

Protected inheritance: In protected inheritance, the public and protected members of the base class become protected members of the derived class, and the private members of the base class remain inaccessible to the derived class.

class Base {
public:
    int publicVar;
protected:
    int protectedVar;
private:
    int privateVar;
};

class Derived : protected Base {
    // publicVar is protected
    // protectedVar is protected
    // privateVar is not accessible
};

The choice of inheritance access specifier depends on the design of the class hierarchy and the intended use of the derived class. Public inheritance is the most common form of inheritance, but private and protected inheritance can be useful in certain situations, such as implementing inheritance for implementation details that should not be exposed to users of a class.

Liskov Substitution Principle

In C++, the Liskov Substitution Principle (LSP) states that if S is a subtype of T, then objects of type T may be replaced with objects of type S without altering any of the desirable properties of the program.

In other words, if a class inherits from another class, it should be possible to use an object of the derived class anywhere an object of the base class is expected without causing any errors or unexpected behavior.

Here is an example that demonstrates the Liskov Substitution Principle:

class Base {
public:
    virtual void print() const { std::cout << "Base" << std::endl; }
};

class Derived : public Base {
public:
    void print() const override { std::cout << "Derived" << std::endl; }
};

int main() {
    Base base;
    base.print();

    Derived derived;
    derived.print();

    Base& ref = derived;
    ref.print();
}

In this example, we define a base class Base with a virtual print() function that prints "Base" to the console. We then define a derived class Derived that inherits from Base and overrides the print() function to print "Derived" to the console.

In the main() function, we create an object base of the Base class and call its print() function. We then create an object derived of the Derived class and call its print() function. Finally, we create a reference ref to the derived object and call its print() function through the reference.

Since the print() function of the Base class is declared as virtual, the print() function of the Derived class is called when the print() function of the base object is called, and the print() function of the Derived class is called when the print() function of the derived object is called. The print() function of the Derived class is also called when the print() function of the base object is called through the reference ref, because the reference ref is bound to the derived object.

If the print() function of the Base class were not declared as virtual, the print() function of the base object would be called when the print() function of the derived object is called through the reference ref, which would cause the program to print "Base" to the console instead of "Derived".
